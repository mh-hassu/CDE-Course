# Grading System

```python
marks = int(input("Enter Your marks:"))
if marks >= 0   and  marks <= 40 :
  print(f"Your Marks are {marks} and Your Grade is F ")
elif marks  >= 40  and marks <= 60 :
  print (f"Your Marks are {marks} and Your Grade is E")
elif marks  >= 60  and marks <= 70 :
  print(f"Your Marks are {marks} and Your Grade is D")
elif marks >= 70  and marks <=80 :
  print(f"Your Marks are {marks} and Your Grade is C")
elif marks >= 80 and marks <=90 : 
  print(f"Your Marks are {marks} and Your Grade is A")
elif marks >= 90 and marks <= 100 :
    print(f"Your Marks are {marks} and Your Grade is A+")
else: 
 print("Invalid Input")
```

# Program For Checking Even or Odd
A program to check whether a number entered by person is even or odd.

```python
try:
 num = int(input("Enter number to check:"))

 if num %2 == 0:
    print(f"{num} is even")
 else:
    print(f"{num} is odd")
except ValueError:
 print("Invalid input. Please Enter a valid input.")
```

# Program that greets you Good morning!

```python
try:
   hour = int(input("Enter the time :"))
   if 0 <= hour <12 :
            print("Good morning!")
   elif 12 <= hour <17:
            print("Good afternoon!")
   elif 17 <= hour <21:
            print("Good evening!")
   elif 21 <= hour <23:
            print("Good night!")
   else:
            print("Invalid hour. PLease enter a hour between 0 to 23 ")
except ValueError:
     print("Invalid Input")
```
