# Lambda Function
- without name
- never been called before or after

```python
(lambda x,y: x+y) (8,9)
```

```python
f = lambda x,y: x*y
```

```python
f(5,8)
```

```python
def f(n):
    if n==1:
        return 1
    s = n + f(n-1)
    return s

f(4)    
```

```python
f = lambda n: 1 if n ==1 else n*f(n-1)
```

```python
f(4)
```

- Also class should have 2 class variables and should dynamic 

```python
class car():
    #class variables
    chasis_no = '1234567'
    engine_type = "Electric"
    # initializer
    def __init__(self,model,color,make_year):
        self.model = model
        self.color = color 
        self.make_year = make_year
        
     # instance method   
    def details(self):
        print(f"Color: {self.color}, Type: {self.model}, Year: {self.make_year}")
     #static method
    @staticmethod
    def price():
        print("Price of Hatchback is around  2 Million")
        
        
    
```

```python
civic = car("Honda","Dark Blue","2019")
```

```python
civic.details()
```
