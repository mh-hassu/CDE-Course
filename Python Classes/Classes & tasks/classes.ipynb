# Classes
- Everything in python belong to some specific **class**
- every class have its **object**
- every class have its own **methods**

```python
name = "Hassaan"
print(type(name))
```

```python
dict = {name : "Hassaan"}
print(type(dict))
```

```python
lst = [1,2,3,5,4]
print(type(lst))
```

```python
class cars():
    def __init__(self,color,model,make_year): # contructor
        self.color = color
        self.model = model
        self.make_year = make_year

    def start(self):   #Method
        """
            start object belongs to class car
            arg: no return
            return: no return
        """

    def stop(self):   # Method
        print("car can stop.")
```

```python
civic = cars('Black','Honda',2025)  # class initialization / object
```

```python
print(type(civic))
```

```python
civic.start()
```

```python
civic.stop()
```

```python
civic.color , civic.model, civic.make_year
```

```python
class human():
    def __init__(self,disability,skin_color):
        self.disability = disability
        self.skin_color = skin_color

    def walk(self):
        print("He can walk.")
    def talk(self):
        print("He can talk.")
```

```python
Ali = human('No','Brown')
```

```python
print(type(Ali))
```

```python
Ali.walk()
```

```python
Ali.talk()
```
