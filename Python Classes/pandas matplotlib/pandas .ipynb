```python
titanic = pd.read_csv('./titanic.csv')
titanic.head()
```

```python
titanic.pivot_table(values='Age',index='Sex',columns='Survived',fill_value=0, margins=True,margins_name='Mean',aggfunc='count')
```

```python
titanic_index = titanic.set_index('Name')
titanic_index.head()
```

```python
titanic_index.reset_index(drop=True).head()
```

```python
titanic['Age'].hist(bins=20)
plt.show
```

```python
gender_count = titanic.groupby('Sex')['Age'].count()
gender_count
```

```python
gender_count.plot(kind = 'bar')
plt.title("World Population in 2007")

plt.xlabel('Sex')
plt.ylabel('Age')

plt.show
```
