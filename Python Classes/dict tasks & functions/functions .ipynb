```python
def fn_bmi(weight,height):
    bmi = weight/(height**2)
    return bmi    
    
```

```python
# calling function
weight =float(input("Enter your weight in kg:"))
height =float(input("Enter your height in m:"))
fn_bmi(weight,height)
```
