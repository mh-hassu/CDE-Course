```python
class Parents():
    def __init__(self,eye_color,hair_color): # constructors
        self.eye_color = eye_color    # --> Attributes
        self.hair_color= hair_color
        
    def can_paint(self):# --> Methods
        print(f"have ability to draw/paint arts")
```

```python
par_obj= Parents('brown','brown')
```

```python
print(par_obj.eye_color)
print(par_obj.hair_color)
par_obj.can_paint()
```

```python
class Child(Parents):
    pass
```

```python
child_obj = Child()
```

```python
child_obj = Child('brown','black')
#-----------------
print(child_obj.eye_color)
print(child_obj.hair_color)
child_obj.can_paint()
```

```python
class Child(Parents):

    def __init__(self,eye_color,hair_color,skin_color):
        super().__init__(eye_color,hair_color)
        self.skin_color = skin_color

    def can_drive(self):
        print(f"Child  can Drive as well.")
```

```python
child_obj = Child('brown','black','brown')
#-----------------
print(child_obj.eye_color)
print(child_obj.hair_color)
print(child_obj.skin_color)
child_obj.can_paint()
```

- create a class with name **Father**which have two attributes and one method
- create a class with name **Child** which inherit characteristics of **Father** but should have atleast **two** more attribute and method other than 'Father'

```python
class Father():
    def __init__(self,height,health):
        self.height = height
        self.health = health
    def can_swim(self):
        print(f"He can swim")

class Mother():
    def __init__(self,beauty,cooking):
        self.beauty = beauty
        self.cooking = cooking
    def cook(self):
        print(f"He can cook")

class Child(Father,Mother):
    def __init__(self,health,height,Knowledge,wealth,beauty,cooking):
        Father.__init__(self,health,height)
        Mother.__init__(self,beauty,cooking)
        self.Knowledge = Knowledge
        self.wealth  = wealth
    def can_drive(self):
        print(f"He can drive.")
```

```python
parent_obj = Father("170 cm","Fit")
```

```python
print(parent_obj.height)
print(parent_obj.health)
parent_obj.can_swim()
```

```python
Mother_obj = Mother("Beautiful","Biryani")
```

```python
print(Mother_obj.beauty)
print(Mother_obj.cooking)
Mother_obj.cook()
```

```python
child_obj = Child("Fit","120 cm","12","lot of money","Beautiful","Karahi")
```

```python
print(child_obj.height)
print(child_obj.health)
child_obj.can_swim()
print(child_obj.beauty)
print(child_obj.cooking)
child_obj.cook()
print(child_obj.Knowledge)
print(child_obj.wealth)
child_obj.can_swim()
```

# OverLoading

```python
from typing import overload

class Adder():
    # ABTRACTION
    @overload
    def add( self , x:int , y:int  ) ->int:
        ...
    @overload
    def add( self , x:float , y:float ) ->float:
        ...
    @overload
    def add( self , x:str , y:str ) ->str:
        ...
    # Actual Functionality
    def add( self , x , y ):
        if isinstance(x, int) and isinstance(y, int):
            return x + y
            
        elif isinstance(x, float) and isinstance(y, float):
            return x + y
        elif isinstance(x, str) and isinstance(y, str):
            return x + y
        else:
            print("Type_Error")
        
    
```

```python
adder_obj = Adder()
```

```python
adder_obj.add(2,5)  # int case
```

```python
adder_obj.add("Saylani","ZA IT PARK") # float case
```

```python
adder_obj.add("Saylani",7)
```
