# ABSTRACTION
- The concept of hiding the complex reality while exposing only the necessary parts.It helps to reduce programming complexity and effort

```python
# abstract class 
class shapes():

    #abstract method
    def sides(self):
        pass
```

```python
class square(shapes):

    def sides(self):
        print("Square has 4 sides")
```

```python
sq_obj = square()
sq_obj.sides()
```

```python
class Circle(shapes):

    def sides(self):
        print("Circle has no sides")
```

```python
cir_obj = Circle()
cir_obj.sides()
```

# Polymorphism
- The ability of different classes to respond to the same message (method call) in different ways.
- This allows for code to work with objects of various classes as if they were objects of a common superclass.

```python
class Animal:
    def speak(self):
        return "Animal Speaks"
class Dog(Animal):
    def speak(self):
        return "woof"
class Cat(Animal):
    def speak(self):
        return "Meow"
# create instances
amm = Animal()
dog = Dog()
cat = Cat()

# call the Speak Method with  on different objects
print(amm.speak())
print(dog.speak())
print(cat.speak())
```

# Data Files
- read --> 'r'
- write --> 'w'
- append --> 'a'
- read/write --> 'rw'

```python
# Method 1 - We have to close file specifically
f = open ('file.txt','w')
f.write('this is content of my file')
f.close()
```

```python
# method 2 - file will close automatically - Recommended
with open ('file.txt', 'w') as f:
    f.write('Now it will overwrite the previous text')
```

```python
with open ('file.txt' , 'a') as f:
    #f.write('Now this line will get appended in the file.\nThis will get appended in file if file exists')
    f.writelines("This is a new line")
```

```python
with open ('file.txt', 'w') as f:
    f.write('Now This will append the previous text')
```

```python
with open ('file.txt', 'r') as f:
    print(f.read())
```
