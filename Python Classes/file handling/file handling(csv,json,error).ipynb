## Csv File

```python
import csv
```

```python
competition = []
year = []
event = []
winner = []

with open ("competition.csv",'r') as f:
    content = csv.reader(f)
    #print(content)
    for line in content:
        competition.append(line)

    print(competition)    
        
       
```

```python

year = []
event = []
winner = []

with open('competition.csv','r') as f:
    content = csv.reader(f)
    #print(content)
    for line in competition[1:]:
        
        year.append(line[0])
        event.append(line[1])
        winner.append(line[2])
        
    print(year,event,winner)
```

```python
with open (file='competition.csv', mode ='a' , newline="") as f :
    write_handler = csv.writer( f , delimiter = ",")
    write_handler.writerows([["2019","World Cup","England"],["2012","Asia Cup","Pakistan"]] )    
```
