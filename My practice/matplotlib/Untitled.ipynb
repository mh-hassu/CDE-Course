# Line PLot

```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
x = [1,2,3,4]
y = [5,6,7,8]

plt.plot(x,y)
plt.title("Line Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.grid()
plt.show()
```

```python
data = { "Salary" : [20000,35000,61000,50000,11115,70000]}
df = pd.DataFrame(data)
df.head()
```

```python
df.shape
```

```python
plt.plot(df["Salary"],color = "red",marker = "o",linestyle = "--",linewidth = '2')
plt.grid()
plt.show
```

# Histogram 

```python
plt.hist(df["Salary"],color = 'brown', density = "light", bins = 5 )
plt.show
```

# Box Plot

```python
plt.boxplot(df["Salary"])
plt.show
```

```python
names = ["Ali", "Hassan", "Adeel", "Saad"]
marks = [80, 65, 90, 75]

plt.bar(names, marks)
plt.title("Students Marks")
plt.xlabel("Name")
plt.ylabel("Marks")

plt.show()
```

```python
plt.scatter([1,2,3,4], [5,3,7,2])
plt.title("Scatter Plot")
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,5))

# 1st plot
plt.subplot(1, 2, 1)
plt.plot([1,2,3], [4,5,6])
plt.title("Plot 1")

# 2nd plot
plt.subplot(1, 2, 2)
plt.bar(["A","B","C"], [3,7,5])
plt.title("Plot 2")

plt.show()
```

```python
sizes = [30, 40, 20, 10]
labels = ["A", "B", "C", "D"]

plt.pie(sizes, labels=labels, autopct="%1.1f%%")
plt.title("Pie Chart")
plt.show()
```

```python
data = [20,22,23,24,25,30,40,45,50]
plt.boxplot(data)
plt.title("Boxplot")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

data = np.random.rand(5,5)

plt.imshow(data, cmap="hot")
plt.colorbar()
plt.title("Heatmap")
plt.show()
```
