# Dictionaries
A dictionary in Python is a collection of key-value pairs.

```python
# Dictionaries    Part 1
my_dict = {
    "name":"Muhammad Hassaan",
    "Degree":"BS CS",
    "Age":"17"
}
print(my_dict)
```

## Accessing Elements

```python
print(my_dict["name"])   # this is use for accessing key of elements
print(my_dict["Degree"])
print(my_dict.get("university")) # safe method   # get is a built in function if the element is in dic it will give but  
                            #  if element is not in that dict like that you want then it will **not** give error
```

### Add & Update values

```python
my_dict["marks"] = 75  # this will add the element in the dict
my_dict["Age"] = 18   # this will update my age
my_dict
```

### Remove elements

```python
students_marks = {"Ali": "80", "Hassaan": "75","Noman":"38","Ahmed":"50","Raza":"92"}
students_marks
```

```python
students_marks.pop("Noman")  # it will take out the elemnt which you want
del students_marks["Ali"] # it is also same , and delete element you want to delete

fruits = {"apple": "1kg","mango":"2kg","Cherry":"200gram"}
fruits.popitem()         #it will always take out the element from last
print(students_marks)
print(fruits)
```

```python
print(students_marks)
students_marks.clear() 
students_marks
```

## Dictionary Methods

```python
customers = {
    "name": "Ali",
    "age" : 17,
    "is_present" : True ,
    "Degree": "CS",
    "number" : "032148363721",
}
customers
```

```python
print(customers.keys()) # all keys will generate
print(customers.values()) # all values of the key will come
print(customers.setdefault("fav_food","biryani"))  # it add new element in the dict,we can also give 1 arg then it will make the value None
print(customers.fromkeys("age",17)) #it make  all alphabets move from each other give equal value,if no value is given taken as None 

print(customers)
print(customers.clear())   # it always clear(Takeout) all the elements of dict 
```

```python
student_01 = {"ahmed":"BBA std"}
student_02 = {"Hassaan":"CS std"}
student_01.update(student_02) # it updates two dictionaries and make changes in one dict
```

# Looping in Dictionaries

```python
months ={"Janurary":1, "Feburaury":2, "March":3, "April":4, "May":5, "June":6, "July":7, "August":8, "September":9, "October":10, 
         "November":11, "December":12}
for temp in months:
    print(temp, ":", months[temp] )
```

```python
for key, value in months.items():
    print(key,"  ",value)
```

# Nested in Dictionaries

```python
school = {
    "student1": {"name": "Ali", "age": 18},
    "student2": {"name": "Ahmed", "age": 19},
    "student3": {"name":"Hassaan","age":17},
    "student4": {"name" : "Raza", "age":15}
}

print(school["student1"]["name"])
print(school["student3"]["age"])
```

# Dictionary Comprehension

```python
squares = {x: x**2 for x in range(1,6)}
print(squares)
```

```python
num = int(input("Enter the table you want to print in dictionary form: "))
table = {temp : (num * temp) for temp in range(1,10)}
print(table)
```

```python
dict_01 = {temp: ("even" if temp %2 ==0 else "odd") for temp in range (1,10)}
print(dict_01)
```
