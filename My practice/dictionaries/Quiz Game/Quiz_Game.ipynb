# Quiz Game

```python
quiz = {
    "What is the capital of Pakistan?":{
        "options":["Karachi","Islamabad","Lahore","Peshawar"], "answer": "Islamabad"
    },
    "Which planet is known as the Red Planet?": {
        "options": ["Earth", "Venus", "Mars", "Jupiter"],
        "answer": "Mars"
    },
    "Who have Discoverd Gravity?": {
        "options" : ["Einstein","Quantum","Maxwell","Newton"],"answer":"Newton"
    }
}
score = 0
for question , data in quiz.items():
    print("\n" + question)
    
    options = data["options"]
    for i, option in enumerate(options, start=1):
        print(f"{i}. {option}")   # Options 1,2,3,4 ke sath
        
    choice = int(input("Enter the answer (1-4):"))
        
    if options[choice - 1].lower() == data["answer"].lower():
        print(" Correct!\n")
        score += 1
    else:
        print(f"Wrong!, Correct Answer is {data["answer"]} ")

        
print(f"\n  Your Final Score is {score}/{len(quiz)}")            
```
