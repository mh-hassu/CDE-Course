# Contact Book

```python
contacts = []
while True :
    print("\n 1. Add Contacts")
    print("2. Show Contacts")
    print("3. Search Contacts")
    print("4. Exit Contact List")

    choice = input("Enter The Choice (1,2,3,4): ")
    if choice == '1':
        name = input("Enter the name: ")
        number = input("Enter the number: ")
        contacts.append([name,number])
    elif choice == '2':
        print("\n ----Contact List----")
        for contact in contacts:
            print(f"Name {contact[0]} | Phone Number {contact[1]}")

    elif choice == '3':
        search_name = input("Enter the name: ")
        found = False
        for contact in contacts:
            if contact[0].lower() == search_name.lower():
                print(f"Found → Name: {contact[0]}, | Phone Number: {contact[1]}")
                found = True 
                break 
        if not found :
            print("Contact not found!")
    elif choice == '4':
        print("\n Exiting Contact list")
        break
    else :
        print("\n Invalid choice! Try again.")
    
```
