# Library Management System

```python
library = {
    "Atomic Habits" : {"author":"James Clear", "copies":3},
    "Think and Grow Rich" : {"author":"Napoleon Hill","copies":5},
    "The 7 Habits of Highly Effective People" : {"author":"Stephen R. Covey", "copies":3},
    "How to Win Friends and Influence People" : {"author":"Dale Carnegie", "copies":7},
    "The Power of Positive Thinking" : {"author":"Norman Vincent Peale", "copies":2},
    "You Can Heal Your Life" : {"author":"Louise Hay", "copies":1},
}
def add_book(title,author,copies):
    if title in library:
        library[title] ["copies"] += copies
        print(f"You borrowed '{title}'. Enjoy reading!")  
    else :
        library[title] = {"author": author , "copies": copies}
def display_book():
    for title, details in library.items():
        print(f"{title} -Author:{details['author']}, copies:{details['copies']}")
def borrowing_book(title):
    if title in library:
        if library[title]['copies'] > 0 :
            library[title]['copies'] -= 1
            print(f"You borrowed {title}. Enjoy reading!")
        else:
            print(f"Sorry, {title} is not available right now.")
            
    else:
        print(f"This book does not exist in the library.")

def returning_book(title):
    if title in library:
        library[title]['copies'] += 1
         print(f"Thank you for returning '{title}' ")
    else:
        print("This book does not belong here")
        
#-------------------------Menu-----------------------------
while True :
    print("\n===== Library Management System =====")
    print("1. Display all books")
    print("2. Add a new book")
    print("3. Borrow a book")
    print("4. Return a book")
    print("5. Exit")
    
    choice = int(input("Enter the number(1-5):"))

    if choice == 1:
        display_book()
    elif choice == 2:
        title = input("Enter the title: ")
        author = input("Enter the author name: ")
        copies = int(input("Enter the copies: "))
        add_book(title,author,copies)
    elif choice == 3:
        title = input("Enter the title: ")
        borrowing_book(title)
    elif choice == 4:
        title = input("Enter the title: ")
        returning_book(title)
    elif choice == 5:
        print("Thank you for using the Library Management System. Goodbye!")
        break
else:
    print("Invalid choice. Please try again.")
```
