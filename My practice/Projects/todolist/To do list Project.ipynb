# To do list Project

```python
# to_do_list = []
while True:
     print("\n 1. Add Task")
     print(" 2. Show Tasks")
     print(" 3. Mark as done")
     print(" 4. Delete Task")
     print(" 5. Exit Task")

     choice = input("Enter The Choice (1,2,3,4,5): ")
     if choice == '1':
         Add_Task = input("Enter the Task: ")
         to_do_list.append(Add_Task)
         
     elif choice == '2':
         print("\n-----To Do List-----")
         
         if not to_do_list:
             print("No Task available.")
         else:
             for temp,task in enumerate(to_do_list, start =1):
                 print(f"{temp}.{task}")
                 
     elif choice == '3' :
         if not to_do_list:
             print("No Task Available.")
         else:
             for temp,task in enumerate(to_do_list,start =1):
                 print(f"{temp}.{task}")
            
         mark = int(input("Enter task number to mark as done: "))
         if 1 <= mark <= len(to_do_list):
            to_do_list[mark-1] = to_do_list[mark-1] + "  (Done)"
            print("Task marked as done!")
         else:
             print("Invalid task number!")
            
     elif choice == '4':
         if not to_do_list:
             print("No tasks to delete!")
         else:
             for i, task in enumerate(to_do_list, start=1):
                 print(f"{i}. {task}")
                 delete = int(input("Enter task number to delete: "))
                 if 1 <= delete <= len(to_do_list):
                     removed = to_do_list.pop(delete-1)
                     print(f"Deleted task: {removed}")
                 else:
                     print("Invalid task number!")
            
     elif choice == '5':
         print("\n Exiting Task")
         break
     else:
         print("\n Invalid choice! Try again.")
```
