# EXTRACTION 

```python
def extract(url,table_attribs):
    ''' ye function website se data uthaye ga jo humein chaiye us ke bd 
       usko data frame me save kre ga ,at the end data frame ko hum neche call kre ge return se
    '''
    
    soup = BeautifulSoup(requests.get(url).text , 'html.parser' )  
    # ye website ko request kre ga phir website se data ko extract kre ga text ki form me or html parse usko html ki tra read kre ga

    table = soup.find('span', string= table_attribs).find_next('table')
    # ye table ko dondhe ga pphir span aik function ha jo group krta ha ye table ko group me badl de ga 
    
    df = pd.readhtml(StringIO(str(table)))[0]
    # ye table ko read kre 
    
    return df
```

```python
import requests
from bs4 import BeautifulSoup

url = 'https://www.ultimatekilimanjaro.com/richest-countries-in-the-world/'
response = requests.get(url)
html_content = response.text

soup = BeautifulSoup(html_content, "html.parser")

# Print Page Title
title = soup.title.string
print(f"Page Title: {title}")

# Extract TABLE
table = soup.find("table")

rows = table.find_all("tr")

print("\nExtracted Table:\n")
for row in rows:
    cols = row.find_all(["th", "td"])
    cols = [c.get_text(strip=True) for c in cols]
    print(cols)
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
from io import StringIO

url = "https://www.ultimatekilimanjaro.com/richest-countries-in-the-world/"

response = requests.get(url)
html_content = response.text

soup = BeautifulSoup(html_content, "html.parser")

# Find table
table = soup.find("table")

# Convert HTML table to DataFrame
html_str = str(table)
df = pd.read_html(StringIO(html_str))[0]

# Save to CSV
df.to_csv("richest_countries.csv", index=False)

print("CSV file created successfully: richest_countries.csv")
```

```python
import requests

def fetch_and_Save_to_file(url,path):
    r = requests.get(url)
    with open(path,"w") as f:
        f.write(r.text)
url = 'https://www.geo.tv/category/pakistan'
fetch_and_Save_to_file(url,"geo_news.html")
```
