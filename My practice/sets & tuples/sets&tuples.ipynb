# What is a Set in Python?
- A set is a collection of different things.

- Written using { } curly brackets, similar to a dictionary but without key-value pairs.


```python
num = {1,2,3,4,5,6} 
print(num)

num_01 = (2,3,7,9,0,8)  # it is also a set
num_01 =set(num_01)
print(type(num_01),num_01)
```

# Key Features of Set
- Unordered → No fixed order.

- Unique → Duplicates are not allowed.

- Mutable → We can add or remove items.

- Heterogeneous → Can store different types (int, float, string, etc.), but not mutable types like lists or dictionaries.

```python
numbr = {10 ,20 ,30 ,40 ,50 ,60 ,70}   # unorder
print(numbr)

num2 = {1,2,4,5,5,1,1,1,3,2,7,3,6}  # donot allow duplicates
print(num2)

my_set = {"Hassaan" ,17 ,65 ,"CS"}  # contain differnt types
print(my_set)
```

### Adding , Removing & Useful operations in Sets

```python
my_set = {"Hassaan" ,17 ,65 ,"CS"}
my_set.add("A+")       # For Adding
print(my_set) 

my_set.remove("CS")  # For Removing
print(my_set)

my_set.pop()    # For Removing last element,As set is unorder so it can take out any element from it
print(my_set)

my_bio = my_set.copy()  # used For copying elements to another variable 
print(my_bio)

fav_sport = {"Football","BasketBall","Cricket"}
indoor_spoorts = {"Chess","Caromboard","Ludo"}
my_set.update(fav_sport,indoor_sports)        # inthis multiple variable can upadate in a sinlge variable 
print(my_set)
```

### Basic Operations in Sets

```python
sports_01 = {"FootBall", "Ludo", "Cricket", "BasketBall", "Hockey","VolleyBall","TableTennis"}

sports_02 = {"BasketBall","TableTennis", "Chess","Caromboard","Ludo"}

print( sports_01.union(sports_02) )   # it gather all elements in single variable

print( sports_01.difference(sports_02) ) # it take out common elements, give resultant from first variable that are left behind
print("\n")
print( sports_01.intersection(sports_02) ) # it gather common elements from both variables

print( sports_01.symmetric_difference(sports_02) ) # it take out common elements, give resultant from both of variables that are left behind
```

```python
A = {1,2,3}
B = {1,2,3,4,5,6,7}
print("A is subset of B:  ",A.issubset(B)) # All elements of A is in B
print("B is subset of A:  ",B.issubset(A)) # ALL elements of B is not in A , there are some common elements

print("\n A is supersubset of B:  ",A.issuperset(B)) 
print("B is supersubset of A:  ",B.issuperset(A))

A.discard(1) # it removes the element from set 
print(A)

B.discard(0)  # if the element isnot in set it willnot give error
print(B)
```

# Tuples
 - A Tuple is a collection of different things but they are immutable(Cannot be changed once it becomes tuple).
 - Written using ( ) round brackets
 - They are in ordered

```python
my_set = ("Hassaan",17,64.5,True,"Computer Science")
print(type(my_set),my_set)

num = {1,2,4,5,9}   # We can also make tuple by {} curly bracket as well
num = tuple(num)
print(type(num),num)
```

### Tuple Methods
 - They have only two methods .index() and .count()

```python
my_set = ("Hassaan",17,64.5,True,"Computer Science")
my_set.index(True)
```

```python
num=(1,2,3,4,5,6,1,1,1,1)
num.count(1)
```

### Nested tuple 

```python
num = ( (1,2,3),4,5,(6,1,1),1,1 )
print(num[3])
print(num[0][2])
```
