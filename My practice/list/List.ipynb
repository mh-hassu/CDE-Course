# 2 .Slicing  
[ start , stop , space ]

```python
numbers = [10, 20, 30, 40, 50]

print(numbers[1:4])   # the ouput will be 10 to 40
print(numbers[:3])    # [10, 20, 30] the output is 10 to 30
print(numbers[2:])    # [30, 40, 50]
print(numbers[::-1])  # [::-1] it reverses the list . if there is [::-2] it reverses the list but then it will leave one element 
                    # it reverses without reverse function
```

# 3 .Adding elements in list

```python
fruits = ["apple", "banana"]

fruits.append("cherry")     # it will Add at the end of the list   & it only add one element
print(fruits)  

fruits.insert(1, "mango" ,)   # it will Add at specific index   & it also add one element
print(fruits) 

fruits.extend(["kiwi", "orange"])  # extend add more than one element
print(fruits)  
```

# 4 .Removing elements from list

```python
fruits = ["apple", "banana", "cherry", "banana"]

fruits.remove("banana")  # it will Removes first 'banana'
print(fruits)  

fruits.pop(1)  # it will Removes element that is at 1 index 
print(fruits)  

last_item = fruits.pop()  # pop always remove element from last
print(last_item)  
print(fruits)

fruits.clear()  # Removes all elements  and make the list empty []
print(fruits)  
```

# Basic Shop 

```python
print("---------- Welcome to Imtiaz Super Market ----------")

cart = []   
while True:
    item = input("Enter the item to add in your cart (type 'done' to finish): ")
    
    if item.lower() == "done":
        break
    cart.append(item)

extra_items = ["Chocolates", "Cold Drink", "Toffees", "Snacks"]
print("\n We also have these items:", extra_items)

choice = input("Do you want to add these items as well? (yes/no): ")

if choice.lower() == "yes":
    cart.extend(extra_items)
    print("\n Your final cart:", cart)
    
elif choice.lower() == "no":
    print("\n Your final cart:", cart)
    
else:
    print("\n Invalid Choice! | Showing cart anyways")
    print("Your final cart:", cart)
```

# Matrix Multiplication

```python
m = int(input("Enter the Rows of Matrix A: "))
n = int(input("Enter the Column of Matrix A: "))
A=[]
for i in range(m):
    row = []
    for j in range(n):
        val = int(input(f"A[{i+1}] [{j+1}]"))
        row.append(val)
    A.append(row)    
print("\n Matrix A: ",A)

o = int(input("\n Enter the Rows of Matrix B: "))
p = int(input("Enter the Column of Matrix B: "))
B = []
for i in range(o):
    row =[]
    for j in range(p):
        val = int(input(f"B[{i+1}][{j+1}]"))
        row.append(val)
    B.append(row)
print("\n Matrix B", B)    

if n != o:
    print("\n Multiplication not  possible.")
else:
    result = [[0 for i in range(p)] for j in range(m)]

    for i in range(m):
        for j in range(p):
            for k in range(n):
                result[i][j] += A[i][k] * B[k][j]

    print("\n Resultant Matrix ")
    for temp in result:
        print(temp)
        
```
