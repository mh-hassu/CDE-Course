```python
def factorial(n):
    if n==0 or n==1:
        return 1
    else:
        return n*factorial(n-1)
        
print(factorial(4)) 
```

```python
global_user = []
def Register():
    name = str(input("Enter the Name: "))
    email = input("Enter the email: ")
    password = input("Enter the password: ")
    global_user.append(email)
    global_user.append(name)
    global_user.append(password)
    print(global_user)
```

```python
Register()
```

```python
def Login(email,password):
    if (global_user[0] == email and global_user[2] == password):
        print(f"Welcome {global_user[1]}")
    else:
        print("Please register first.")
```

```python
Login("hassu@gmail.com","12345")
```

```python
Login("ahmed@gmail.com","12345")
```

```python
def display_result(winner, score,**other_info):
    print("The Winner was "+ winner)
    print("The Score was "+ score)
    for key, value in other_info.items():
        print(key ,":",value)
```

```python
display_result(winner="Hassaan",score = "3-0", halftime = "2",injuries = 'No')
```
