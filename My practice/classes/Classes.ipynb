```python
class Car():  
    def __init__(self,brand,model,color):
        self.brand = brand
        self.model = model
        self.color = color
    def show(self):
        print(f"Car: {self.brand} , {self.color} , {self.model}")
        
        
```

```python
car1= Car("Honda","2019","Dark Blue")
```

```python
print(car1.brand)
print(car1.model)
print(car1.color)
car1.show()
```

## Inheritence 

```python
class Father():
    def __init__(self,height,health,skin_color):
        self.health = health
        self.height = height
        self.skin_color = skin_color
    def can_drive(self):
        print("He can Drive")
class Child(Father):   
    def __init__(self,eye_color,hair_color,height,health,skin_color):
        super().__init__(height,health,skin_color)
        self.eye_color = eye_color
        self.hair_color = hair_color
    def can_paint(self):
        print(f"have ability to draw/paint arts")
```

```python
father_obj = Father("6 ft","good","brown")
```

```python
print(father_obj.health)
father_obj.can_drive()
```

```python
child_obj = Child("Black","White","5.8 ft","fit","Brown")
```

```python
print(child_obj.skin_color)
print(child_obj.height)
child_obj.can_paint()
```
